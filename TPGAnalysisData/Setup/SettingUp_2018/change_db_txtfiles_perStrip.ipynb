```python
#Reads in the db file and the per ring stripid enc weights file
#finds the stripids of all crystals in ring 28 for example
#changes the db text file weight groups of the stripids in ring 28

#can find all stripids in ring 28 for example and change the weight group associated with them. good!
import pandas as pd
import numpy as np

data = pd.read_csv("/eos/user/w/wrsmith/SWAN_projects/NEW_March/SettingUp/DOF_P_Jun18_5_1_20Nov_with_stripID_encw_NEW.txt",  sep='\t', float_precision='round_trip')
txtfile = pd.read_csv("/eos/user/w/wrsmith/SWAN_projects/Load_db_weights/Per_Strip/EcalTPGWeightGroup_original.txt", delim_whitespace=True, header=None)

wgroups = []
stripidlist = []

for q in data.index:
    
    
    weights = [data.loc[q, 'encw1'], data.loc[q, 'encw2'], data.loc[q, 'encw3'], data.loc[q, 'encw4'], data.loc[q, 'encw5']]
    #print(weights)
    
    stripid = data.loc[q, 'stripid']
    #print(etaring)
            
    if stripid not in stripidlist:
        #print(weights)
        stripidlist.append(stripid)

        wgroups.append(weights)

print("\n", stripidlist)
print("\n", len(stripidlist))
print("\n", wgroups)
print("\n", len(wgroups))
            
```

```python
   
for group, s in enumerate(stripidlist):

    for j in txtfile.index:
        #print(j)
        #Must add 1 to the weight groups as wgroup[0] is for EB strips
        #f = group+1
        if txtfile.loc[j, 0] == s:
            txtfile.at[j, 1] = group
            
#print(txtfile.loc[(txtfile[1] == 3)])   

txtfile.to_csv("./EcalTPGWeightGroupALL.txt", sep=' ', index=False, header=None)

print("finished")    
    
```

```python
#Creates Weight Id MAP textfile per ring

groupsTXT = pd.DataFrame(wgroups)


groupsTXT = groupsTXT.apply(np.int64)
#print(groupsTXT)

groupsTXT.to_csv("./EcalTPGWeightIdMAPALL.txt", sep=' ', index=False, header=None)
print("finished")
```
