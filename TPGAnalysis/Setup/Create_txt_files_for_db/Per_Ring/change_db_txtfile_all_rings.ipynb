```python
#Reads in the db file and the per ring stripid enc weights file
#finds the stripids of all crystals in ring 28 for example
#changes the db text file weight groups of the stripids in ring 28

#can find all stripids in ring 28 for example and change the weight group associated with them. good!
import pandas as pd
import numpy as np

data = pd.read_csv("/eos/user/w/wrsmith/SWAN_projects/Data/DOF_encWeights_perRING_Jun18_6_stripid.txt",  sep='\t', float_precision='round_trip')
txtfile = pd.read_csv("/eos/user/w/wrsmith/SWAN_projects/Load_db_weights/EcalTPGWeightGroup_original.txt", delim_whitespace=True, header=None)

wgroups_pos = [0]*29
wgroups_neg = [0]*29
etaringlist = []

for q in data.index:
    
    
    weights = [data.loc[q, 'encw1'], data.loc[q, 'encw2'], data.loc[q, 'encw3'], data.loc[q, 'encw4'], data.loc[q, 'encw5']]
    #print(weights)
    
    etaring = data.loc[q, 'eta_ring']
    #print(etaring)
    
    #if etaring == 0:
        #print(data.loc[q, 'CMSSWID'])
        
    if etaring != 0:
        
        if etaring not in etaringlist:
            #print(weights)
            etaringlist.append(etaring)

            if etaring > 0:
                wgroups_pos[etaring] = weights
            else:
                wgroups_neg[abs(etaring)] = weights 

print("\n", etaringlist)
print("\n", len(etaringlist))
print("\n", wgroups_pos)
print("\n", wgroups_neg)








            
```

```python
for ring in etaringlist:
    
    print(ring)
    strips = []
    
    for i in data.index:
    
        if data.loc[i, 'eta_ring'] == ring:
            stripid = data.loc[i,'stripid']
            strips.append(stripid)
            #print(data.loc[[i]])
    

    strips = np.unique(strips)        
    #print(strips)
    #print(len(strips))
        
    
    group = etaringlist.index(ring)
    
    for s in strips:
        #print(s)
    
        for j in txtfile.index:
            #print(j)
            if txtfile.loc[j, 0] == s:
                txtfile.at[j, 1] = group
            
#print(txtfile.loc[(txtfile[1] == 3)])   







txtfile.to_csv("/eos/user/w/wrsmith/SWAN_projects/Load_db_weights/EcalTPGWeightGroupALL.txt", sep=' ', index=False, header=None)
    
    
print("finished")    
    
```

```python
groupstxtfile = []

for g in etaringlist:
    if g < 0:
        groupstxtfile.append(wgroups_neg[abs(g)])
    else:
        groupstxtfile.append(wgroups_pos[g])
        
print(groupstxtfile)
    
```

```python
#Creates Weight Id MAP textfile per ring

groupsTXT = pd.DataFrame(groupstxtfile)


groupsTXT = groupsTXT.apply(np.int64)
print(groupsTXT)

groupsTXT.to_csv("/eos/user/w/wrsmith/SWAN_projects/Load_db_weights/EcalTPGWeightIdMAPALL.txt", sep=' ', index=False, header=None)
print("finished")
```
